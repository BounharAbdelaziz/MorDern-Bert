## Load Dataset

```python
from datasets import load_dataset
from google.colab import userdata
```

```python
token=userdata.get("hf_read")
ds=load_dataset("atlasia/AL-Atlas-Moroccan-Darija-Pretraining-Dataset",token=token)
ds
```

```python
ds["train"].remove_columns(["metadata","dataset_source"])
```

```python
train_ds=ds["train"].remove_columns(["metadata","dataset_source"]).select(range(300_000))
test_ds=ds["test"].remove_columns(["metadata","dataset_source"])
```

```python
def process_function(examples):
  return tokenizer(examples["text"],truncation=True)
```

```python
tokenized_train=train_ds.map(process_function,batched=True)
tokenized_test=test_ds.map(process_function,batched=True)
```

```python
tokenized_train
```

```python
# data collator: mask the whole of word
from transformers import DataCollatorForLanguageModeling
data_collator=DataCollatorForLanguageModeling(
    tokenizer
)
```

```python
# trainer
from transformers import TrainingArguments,Trainer
args=TrainingArguments(
    "model",
    learning_rate=2e-5,
    per_device_eval_batch_size=8,
    per_device_train_batch_size=4,
    num_train_epochs=1,
    save_total_limit=2,
    weight_decay=0.01,
    evaluation_strategy="steps",
    eval_steps=1000,
    logging_steps=1000,
    warmup_ratio=0.03,
    fp16=True,
    gradient_checkpointing=True,
    gradient_accumulation_steps=4,
    report_to="wandb",
    run_name="Bert Continuos Training"
)
```

```python
trainer=Trainer(
    model=model,
    args=args,
    train_dataset=tokenized_train,
    eval_dataset=tokenized_test,
    tokenizer=tokenizer,
    data_collator=data_collator
)
```

```python
trainer.train()
```
